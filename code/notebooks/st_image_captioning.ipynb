```python
from transformers import BlipProcessor, BlipForConditionalGeneration
from PIL import Image

class ImageCaptionGenerator:
    # default processor and model
    # BLIP (Bootstrapped Language-Image Pre-training) from HuggingFaces
    def __init__(self):
        self.processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-base")
        self.model = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-base")

    # pass the custom/other processor and model
    def custom(self, processor, model):
        self.processor = processor
        self.model = model

    def generate_caption(self, image_path):
        image = Image.open(image_path)
        inputs = self.processor(images=image, return_tensors="pt")
        caption = self.model.generate(**inputs)
        caption_text = self.processor.decode(caption[0], skip_special_tokens=True)
        return caption_text
```
