## Convert an image to a b64 string

```python
import base64

with open('../../assets/sofa.jpg', "rb") as f:
    image_str = base64.b64encode(f.read()).decode("utf-8")
    print(image_str)
```

```python
from IPython.display import display
from PIL import Image
import io

display(Image.open(io.BytesIO(base64.b64decode(image_str))))
```
